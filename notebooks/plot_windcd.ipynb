```python
import netCDF4
from datetime import datetime, timezone
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = "/home/runnalja/git/alplakes/simulations/runs/eawag_delft3dflow6030062434_delft3dflow_geneva_20240128_20240228_hourly/trim-Simulation_Web.nc"
with netCDF4.Dataset(file) as nc:
    for i in range(nc.variables["WINDCD"].shape[0]):
        data = np.array(nc.variables["WINDCD"][i,:])
        data[data == -999] = np.nan
        t = nc.variables["time"][i]
        t = datetime.utcfromtimestamp(t + (datetime(2008, 3, 1).replace(tzinfo=timezone.utc) - datetime(1970, 1, 1).replace(tzinfo=timezone.utc)).total_seconds()).replace(tzinfo=timezone.utc)
        print(t)
        plt.imshow(data, cmap='coolwarm', interpolation='nearest')
        plt.colorbar()
        plt.show()
```
